## Finding assets that have been oversold

* get drawdown since peak as percentage
* compare as group

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 12
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size
```

```python
df = pd.read_csv('/home/ian/Downloads/etfs_details_type_fund_flow.csv')
df.drop(columns=[ 'Asset Class','YTD','Avg Volume', 'Previous Closing Price', 'Inverse', 'Leveraged',
       '1 Week', '1 year', '3 year', '5 year',
       'Inception', 'ER', 'Commission Free',
       'Dividend Date', 'Dividend', 'Annual Dividend Yield ', 'P/E Ratio',
       'Beta', '# of Holdings', '% In Top 10', 'Overall Rating',
       'Liquidity Rating', 'Expenses Rating', 'Returns Rating',
       'Volatility Rating', 'Dividend Rating', 'Concentration Rating',
       'ESG Score'],inplace=True)
```

```python
df=df[df['Total Assets ']>150000000]
df=df[df['Annual Dividend Rate']>0]
df=df[df['4 Week']<0]
#df.sort_values(by=['Total Assets ','4 Week'], ascending=False)
```

```python
fig, axs = plt.subplots(sharey=True)
names = list(df['4 Week'])
values = list(df['ETFdb.com Category'])
axs.scatter(names,values)
```
